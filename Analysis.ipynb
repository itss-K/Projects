# Adding  column 'Sales' and 'Month'. 

```python
all_data['Month']=all_data['Order Date'].str[0:2]

all_data.tail(20)
```

```python
#coverting column to right dtype

all_data['Price Each']=pd.to_numeric(all_data['Price Each'])
all_data['Quantity Ordered']=pd.to_numeric(all_data['Quantity Ordered'])
```

##### city having the highest no of sales

```python
#adding a column 'city'
#using apply method

def get_city(add):
    return add.split(',')[1]

all_data['City']=all_data['Purchase Address'].apply(lambda x: get_city(x) )

def get_state(add):
    return add.split(',')[2].split(' ')[1]

all_data['State']=all_data['Purchase Address'].apply(lambda x:get_state(x))

all_data.head(20)

```

## Products most often brought together 

```python
all_data.head(15)
```

```python
df=all_data[all_data['Order ID'].duplicated(keep=False)]
df.head()
```

```python
df['Grouped']=all_data.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
df.head()
```

```python
df=df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

count=Counter()

for row in df['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
#pair of 2
for k in count.most_common(10):
    print(k)

```

```python
from itertools import combinations
from collections import Counter

count=Counter()

for row in df['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,3)))
    
#pair of 3
for k in count.most_common(10):
    print(k)

```

```python
from itertools import combinations
from collections import Counter

count=Counter()

for row in df['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,4)))
    
#pair of 4
for k in count.most_common(10):
    print(k)

```
